```python
import boto3
import json

# Create the Bedrock runtime client
client = boto3.client("bedrock-runtime")

# Build a payload in the correct Nova Lite format
payload = {
    "schemaVersion": "messages-v1",  # must be included
    "messages": [
        {
            "role": "user",              # only "user" or "assistant"
            "content": [
                {"text": "Explain what a hello world program does."}  # content must be array of objects with text
            ]
        }
    ]
}

# Call the model
response = client.invoke_model(
    modelId="amazon.nova-lite-v1:0",  # your active Nova Lite model
    body=json.dumps(payload),
    contentType="application/json",
    accept="application/json"
)

# Parse and inspect the output
result = json.loads(response["body"].read())

print(json.dumps(result, indent=2))
```
