```python
from langchain_aws import ChatBedrock
import os

llm = ChatBedrock(
    model_id="amazon.nova-micro-v1:0",
    region_name=os.environ["AWS_REGION"]
)

response = llm.invoke([
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"}
])

print(response)
```
